```python
import matplotlib.pyplot as plt
import math
```

```python
%matplotlib inline
```

```python
ta =[]
tao=[]
def TA(n):
    return 1000*n
def TAO(n):
    return n
for i in range(1,17):
    tao.append(TAO(i))
for i in range(1,17):
    ta.append(TA(i))

plt.plot(ta) 
```

```python
tb =[]
tbo =[]
def TB(n):
    return 200*(n*math.log(n))

def TBO(n):
    return (n*math.log(n))


for i in range(1,17):
   tbo.append(TBO(i))

for i in range(1,17):
   tb.append(TB(i))

plt.plot(tb) 
```

```python
tc =[]
tco =[]
def TC(n):
    return 10*(n**2)

def TCO(n):
    return 10*(n**2)

for i in range(1,17):
    tco.append(TCO(i))

for i in range(1,17):
    tc.append(TC(i))

plt.plot(tc) 
```

```python
td =[]
def TD(n):
    return 2**n

for i in range(1,17):
    td.append(TD(i))

plt.plot(td) 
```

```python
plt.plot(ta, label="1000n")
plt.plot(tb, label="200nlog(n)")
plt.plot(tc, label="10n^2")
plt.plot(td, label="2^n")
plt.legend()
plt.xlabel("n")
plt.ylabel("f(n)")
plt.title("Complexities")
plt.show
```

```python
plt.plot(tao, label="n")
plt.plot(tbo, label="nlog(n)")
plt.plot(tco, label="n^2")
plt.plot(td, label="2^n")
plt.legend()
plt.xlabel("n")
plt.ylabel("f(n)")
plt.title("Complexities")
plt.show
```
